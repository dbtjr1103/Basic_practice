### 35장 정규표현식 


- 정규표현식
    - 어떤 문자열에서 내가 원하는 패턴을 찾는 기능
    - 모든 프로그래밍 언어에서 제공하는 어떤 문자열을 다룰때 꼭 써야하는 기능 중 하나 
    - 메타문자

    ```
     +, (), [], / 등 문자가 특별한 용도로 사용되는 문자
     [] : [] 사이의 문자들과 매치
        ex) [abc]라면 a,b,c 중 한 개의 문자와 매치, [a-c] 표현 가능
     [0-9] : 0부터 9까지 모든 숫자
     [a-zA-Z] : 영어
     [가-힣] : 한글
     [0-9a-zA-Z가-힣] : 전부다
    ```
    - 문자 클래스
    ```
    \d : 숫자
    \D : 숫자가 아닌 것
    \s : 공백
    \S : 공백이 아닌 것
    \w : 문자 + 숫자
    \W : 문자 + 숫자 아닌 것
    ```      
    - 점(.)
    ```
    \n을 제외한 모든 문자와 매치
        ex)a.b라면 aab, a0b등 모두 가능
    ```      
    - 반복(*)
    ```
    0부터 무한대로 반복될 수 있다는 뜻
        ex) ca*t
        ct, cat, caaaaat 등
    ```      
    - 반복(+)
    ```
    1부터 무한대로 반복될 수 있다는 뜻
        ex) ca*t
        cat, caat 등
    ```      
    - 반복 {}
    ```
    {n} : 반드시 n번 반복
        ex) ca{3}t -> caaat
    {m, n} : m~n번 반복
        ex) ca{1,3}t -> cat, caat, caaat
    ```    
    - ?
    ```
    있어도 되고 없어도 된다라는 의미
        ex) ab?c
        ac, abc
    ```    
    - re
    ```
    정규표현식 라이브러리
    import re

    re.findall(패턴, 문자열)
     ```
