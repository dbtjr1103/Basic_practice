문제 11) 타이타닉 데이터프레임에 "성별2" 열을 추가하세요.
* "성별2" 열은 "남자", 여자"로 구성되지만, 나이가 20살 미만이라면 "어린이" 라는 값으로 값을 구성해주세요.

import pandas as pd

titanic = pd.read_csv("titanic.csv")

```python
titanic['성별2'] = titanic['sex']

titanic['성별2'] = titanic['성별2'].replace({'male': '남자', 'female': '여자'})

titanic.loc[titanic['age'] < 20, '성별2'] = '어린이'
titanic
```

* "생존" 열은 survived 열입니다.

import pandas as pd

import seaborn as sns


titanic = pd.read_csv("titanic.csv")

```python
df = abs(titanic.corr()) # 상관관계 데이터프레임 생성 및 절댓값 씌우기

df2 = df.loc[(df['survived'] != 1),["survived"]] # 생존열 추출 및 상관계수가 1인행 제거

df2.idxmax()[0] # 상관관계가 가장 높은 열 이름 추출
```
