# 26장 seaborn 라이브러리를 활용해서 그래프 그리기

- seaborn

    ```
    import seaborn as sns
    ```
    - 장점 : 자동으로 예쁜 그림을 그려준다
    - 단점 : 커스터마이징이 힘들다
- seaborn과 matplotlib는 연동이 가능하다

    ```
    fig = plt.figure()
    ax1 = fig.add_subplot(1,1,1)
    ax1 = sns.histplot(tips['total_bill'])
    ```    
- 히스토그램
    ```
    sns.histplot(데이터, bins = 막대기개수)
    단변량 그래프, 일변량 그래프
    ```     
- 밀도 함수 그래프
    ```
    sns.kdeplot(데이터)
    sns.kdeplot(데이터1, 데이터2)
    단변량 그래프
    2차원부터는 : 이변량 그래프
    ```   
- 특정 값 개수 그래프
    ```
    sns.countplot(데이터)
    단변량 그래프
    ```   
- 산점도
    ```
    sns.lmplot(x='total_bill', y='tip', data=tips)
    이변량 그래프
    ```   
- 산점도 + 히스토그램
    ```
    sns.jointplot(x='total_bill', y='tip', data=tips)
    이변량 그래프
    ```   
- 평균 막대 그래프
    ```
    sns.barplot(x='time', y='total_bill', data=tips)
    이변량 그래프
    ```       
- 상자 그래프
    ```
    sns.boxplot(x='time', y='total_bill', data=tips)
    이변량 그래프
    ```       
- 바이올린 그래프
    ```
    sns.violinplot(x='time', y='total_bill', data=tips)
    이변량 그래프
    ```       
- 관계 그래프
    ```
    sns.pairplot(tips)
    ```       
- 다변량 그래프 그래프
    ```
    매개변수 hue : 색상 추가

    ```
    sns.set_style(배경) : 그래프 배경 설정

# 28장 여러 개의 데이터 프레임을 합치는 방법

- concat
    : 여러개의 데이터 프레임을 하나의 데이터프레임으로 합치는 함수

- 열 이름을 기준으로 합침

    ```
    pd.concat([df1, df2, ...])
    ```      
- 행 이름을 기준으로 합치기
    ```
    pd.concat([df1, df2, ...], axis = 1)
    axis(축) : 1일때 열방향, 0일때 행방향
    ```   
- 행 이름 바꾸기
    ```
    df.index = [...]
    ```   
- 열 이름 바꾸기
    ```
    dv.columns = [...]
    ```

# 29장 스마트하게 데이터 프레임을 합치는 방법

- merge : 특정 열을 기준으로 데이터 프레임을 합치는 방법

    ```
    df1.merge(df2, left_on=df1의 기준열, right_on=df2의 기준열)
    기준열의 이름이 서로 같은 경우 on으로 대처가능
    ```      

````python
import pandas as pd
````

````python
# 날씨 정보
person = pd.read_csv("survey_person.csv")
site = pd.read_csv("survey_site.csv")
survey = pd.read_csv("survey_survey.csv")
visited = pd.read_csv("survey_visited.csv")
````

````python
person
````

````python
site
````

````python
survey
````

````python
visited
````

````python
# visited와 site merge
merge1 = visited.merge(site, left_on = "site", right_on="name")
del merge1['name'] #중복열 삭제
merge1
````

````python
merge2 = survey.merge(person, left_on='person', right_on='ident')
del merge2['ident']
merge2
````

````python
merge3 = merge2.merge(merge1, left_on='taken', right_on='ident')
del merge3['ident']
merge3
````

# 30장 누락값을 처리하는 방법

- 누락값
    - 누락값이란 비어있는 값을 말함
    - 누락값이 하나라도 있다면 데이터분석 및 연산이 원활하지 못할 수 있음
    - 누락값은 비교할 수 없음 단, pd.isnull() 함수로 누락값인지 아닌지 판단가능

    ```
    from numpy import NAN, NaN, nan
    ```      
    ```
    pd.isnull() : 누락값이면 True, 아니면 False
    pd.notnull() : 누락값이 아니면 True, 누락값이면 False
    ```      
- 누락값 개수 구하기
    - 데이터 개수 - df.count()
    ```
    np.count_nonzero(데이터.isnull())
    df['Cases_Guinea'].value_counts(dropna=False)
    ```
    - 누락값 채우기 : fillna(값)
    ```
    fillna(method = 'ffill') -> front fill
    fillna(method = 'bfill') -> back fill
    df.interpolate() -> (선형)흐름에 따라 결측치를 채워줌
    ```      
    


````python
#contry_timeseries.csv 에볼라 바이러스 데이터
import pandas as pd

ebola = pd.read_csv("country_timeseries.csv")
ebola
````

````python
# rows, colums 확인
ebola.info()
````

````python
# 누락값 
122 - ebola.count()
````

````python
ebola.fillna(5)
````

````python
ebola.mean() #평균값 확인
````

````python
ebola.fillna(ebola.mean())
# 시계열 데이터에는 결측치를 평균값으로 채우면 안됨
````

````python
ebola.fillna(method = 'ffill')
# 앞의 값이 없어서 결측치가 있음
````

````python
ebola.fillna(method='bfill')
# 뒤의 값이 없어서 결측치가 있음
````

````python
# 흐름에 따라 누락값을 채우는 기능
ebola.interpolate()
````

````python
ebola2 = ebola.interpolate()
ebola2 = ebola2.fillna(method='ffill')
ebola2 = ebola2.fillna(method='bfill')
ebola2
````
