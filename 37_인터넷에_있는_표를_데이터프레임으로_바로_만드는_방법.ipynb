# 37장 인터넷에 있는 표를 데이터프레임으로 바로 만드는 방법


- 표 수집
    - 수집한 표 정보를 바로 표(데이터프레임)으로 변환할 수 있다.
    ```
    pd.read_html(str(HTML))
    표 데이터만 크롤링 해서 넣어주면 됨
    html.find를 써서 표를 찾아서 문자열(str)로 넣어주어야 함

    ```

````python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
````

````python
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1")
html = BeautifulSoup(url.text)
kospi_page = int(html.find('td', class_ = 'pgRR').find('a')['href'].split('=')[-1])

total = []

for n in range(1, kospi_page + 1):

    url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={}".format(n))
    html = BeautifulSoup(url.text)
    
    table = html.find('table', class_ = 'type_2')
    table = pd.read_html(str(table))[0] # 리스트 안에 있어서 [0]
    # NaN값들 - 실선, 등 
    table = table[table['종목명'].notnull()]

    del table['N']
    del table['토론실']

    total.append(table)
    time.sleep(1)

    print("{}페이지 저장완료...".format(len(total)))
````

````python
# 모든 코스피 페이지
kospi_page = int(html.find('td', class_ = 'pgRR').find('a')['href'].split('=')[-1])
````

````python
# 여러개의 df를 하나의 프레임으로 변경
kospi = pd.concat(total, ignore_index=True)
kospi.to_excel("kospi.xlsx")
kospi
````
