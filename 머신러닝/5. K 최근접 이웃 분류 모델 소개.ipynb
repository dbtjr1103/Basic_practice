# 5장 K 최근접 이웃 분류 모델 소개

- 가장 가까운 데이터를 탐색해서 값을 예측하는 모델
    - 새로운 입력으로 들어온 데이터를 특정값으로 분류하는데 현재 데이터와

- 단점
    - 속도가 아주 느리다
    - 하나의 예측을 분류하기 위해서 저장된 모든 데이터와 비교해야함
    - 적절한 K값을 찾아야 함
    - 데이터베이스에 모든 데이터가 저장되어야 함(속도 저하) 

```python
import pandas as pd

fish = pd.read_csv("fish.csv")
fish
```

```python
import seaborn as sns

sns.lmplot(x='length', y='weight', data=fish, hue='class', fit_reg=False)
```
