```python
import requests
from bs4 import BeautifulSoup

url = 'https://finance.naver.com/sise/sise_index_day.naver?code=KPI200&page=1'

resp = requests.get(url)
# resp
soup = BeautifulSoup(resp.content, 'lxml')
pg_tags = soup.select('td.pgRR a')
last = int(pg_tags[0]['href'].split('=')[-1])
last # 마지막 번호


from tqdm.notebook import tqdm

dates, prices, ratios, rates, volumes, amounts = [], [], [], [], [], []

for pg in tqdm(range(1,last+1)):
    url = f'https://finance.naver.com/sise/sise_index_day.naver?code=KPI200&page={pg}'

    resp = requests.get(url)
    soup = BeautifulSoup(resp.content, 'lxml')
    date_tags = soup.select('td.date')
    ratio_tags = soup.select('td.rate_down span')
    price_tags = soup.select('td.number_1')[::4]
    rate_tags = soup.select('td.number_1')[1::4]
    volume_tags = soup.select('td.number_1')[2::4]
    amount_tags = soup.select('td.number_1')[3::4]


    for i in range(len(date_tags)):
        dates.append(date_tags[i].text)
        prices.append(price_tags[i].text)
        if ratio_tags[i]['class'][-1]  == 'nv01':
            ratios.append('-' + ratio_tags[i].text.strip())
        else:
            ratios.append(ratio_tags[i].text.strip())
        rates.append(rate_tags[i].text.strip())
        volumes.append(volume_tags[i].text)
        amounts.append(amount_tags[i].text)

kpi = {
    '날짜' : dates,
    '체결가' : prices,
    '전일비' : ratios,
    '등락률' : rates,
    '거래량' : volumes,
    '거래대금' : amounts
}

# kpi
```

```python
import pandas as pd

df_kpi = pd.DataFrame(kpi)
```
